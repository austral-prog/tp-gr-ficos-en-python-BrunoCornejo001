```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.normal(loc=0, scale=1, size=1000)
import os
os.makedirs('imgs', exist_ok=True)
```

# Velocidad de un auto

## Velocidad de un auto en tres tramos distintos de calle. Cuya velocidad en m/s va aumentando.

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro-')
plt.savefig('imgs/line_plot.png', dpi=300, bbox_inches='tight')
```

# Votos elecciones

## A = Votos de Bullrich, B = Votos de Milei, C = Votos Massa

```python
plt.bar(['A','B','C'], [3,7,2])
plt.savefig('imgs/bar_plot.png', dpi=300, bbox_inches='tight')
```

# Crecimiento Poblacional de Carpinchos

## Poblacion de carpinchos a lo largo del siglo pasado en Argnetina

```python
plt.plot(np.random.random(100).cumsum())
plt.savefig('imgs/custom_plot.png', dpi=300, bbox_inches='tight')
```

# Edades

## Frecuencia de edades de personas que reunieron a jugar al bingo en la feria etatal

```python
plt.hist(data, bins=30, color="skyblue", edgecolor="black", alpha=0.7)
plt.title("Histograma de los datos")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
plt.savefig('imgs/histogram.png', dpi=300, bbox_inches='tight')
```

# Porciones de torta

## Porcentaje de torta que le toca a mi hermana (20%), Porcentaje de torta que le toca a mi perro (30%), porcentaje de torta que me toca a mi (50%)

```python
valores = [30, 20, 50]
etiquetas = ['A', 'B', 'C']

plt.pie(valores, labels=etiquetas, autopct='%1.1f%%', startangle=90)
plt.title("Distribucin de categoras")
plt.axis('equal')  # Para que sea circular
plt.show()
plt.savefig('imgs/pie_plot.png', dpi=300, bbox_inches='tight')
```
